```python
##opencv- open computer vision
import cv2
frame=cv2.imread("yo.png")## returns BGR image 
plt.imshow(frame)
```

```python
##converting bgr to rgb
rgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
# "haarcascade_frontalface_default.xml" this is basically a ml trained model that detects the face
rgb1=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
faces=classifier.detectMultiScale(rgb1)
print(faces)
face=faces[0]
x,y,w,h=face
face=rgb1[y:y+h,x:x+w]
plt.imshow(face)
```

```python
output=cv2.rectangle(rgb1,(x,y),(x+w,y+h),(0,0,255),4)
plt.imshow(output)
```

```python
out=cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),10000)
plt.imshow(out)
```

```python
import cv2
cap=cv2.VideoCapture(0)
classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
# "haarcascade_frontalface_default.xml" this is basically a ml trained model that detects the face
while True:
    ret,frame=cap.read()
    if ret:
        faces=classifier.detectMultiScale(frame)
        for face in faces:
            x,y,w,h=face
            
            frame=cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),4)
        cv2.imshow("THIS IS ME",frame)
    key=cv2.waitKey(1)#so here what we are doing that we are waiting for the key to be pressed by the user after every 1 sec so if you 
                        #increse the time to say 600 so in between 600 sec if you press a key then it wont take ypur input, that is it will 
                    #take a image then wait for 600 sec then another then again wait for 600 sec so the image will appear laggy.
    if key==ord("r"):
        break
cap.release()
cv2.destroyAllWindows()
```
